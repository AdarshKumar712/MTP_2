```python
#defining function to read MIDI files
def read_midi(file):
    
    print("Loading Music File:",file)
    
    notes = np.load(file, allow_pickle=True)

    for i in range(len(notes)):
      notes[i]= np.array(notes[i])

    return np.array(notes)
```

```python
from miditoolkit import MidiFile
from miditok import MIDILikeEncoding, get_midi_programs
text = '[CLS]'
pitch_range = range(21, 109)
beat_res = {(0, 4): 8, (4, 12): 4}
nb_velocities = 32
additional_tokens = {'Chord': True, 'Rest': True, 'Tempo': True, 'Program': False,
                     'rest_range': (2, 8),  # (half, 8 beats)
                     'nb_tempos': 32,  # nb of tempo bins
                     'tempo_range': (40, 250),
                     'TimeSignature':False}  # (min, max)
tokenizer_midi = MIDILikeEncoding(pitch_range, beat_res, nb_velocities, additional_tokens)
```

```python
from tqdm.notebook import tqdm
```

```python
for i in tqdm(range(20)):
  predicted_sequences = generate_music(i, Notes_len=800)
  predicted_sequences_ign = []
  for j in range(len(predicted_sequences)):
    if predicted_sequences[j]>=331:
      continue
    predicted_sequences_ign.append(predicted_sequences[j])
  save_name = "./base/generated/" + "sample_"+str(i+1)+"_800.mid"
  gen_midi = tokenizer_midi.tokens_to_midi([predicted_sequences_ign], [(0, False)])
  gen_midi.dump(save_name)
```

```python
#for listing down the file names
import os

#Array Processing
import numpy as np

#reading each midi file
notes_array = read_midi("./../data/test_bpe.npy")
```

```python
def generate_music(ind, Notes_len=200):
    random_music =  np.array([notes_array[ind][:200]])
    config.length = Notes_len
    inputs = torch.from_numpy(random_music)
    result = mt(inputs, config.length, gen_summary_writer)
    return result
```

```python
def convert_to_midi_bpe(predicted_sequences, file_save="sample.mid"):
  s1 = tokenizer_nlp.decode(predicted_sequences)
  s1 = "".join(s1.split(' '))
  tokens_real = []
  for i in range(len(s1)):
      tokens_real.append(int(ord(s1[i])-int(0x2200)))
  print("Predicted Sequence Length: ", len(predicted_sequences))
  print("Midi token length Remi: ", len(tokens_real))
  
  gen_midi = tokenizer_midi.tokens_to_midi([tokens_real], [(0, False)])
  gen_midi.dump(file_save)
  print("Midi saved successfully")
  print()
```
