```python
from miditok import REMIEncoding, get_midi_programs
from miditoolkit import MidiFile
```

```python
text = '[CLS]'
pitch_range = range(21, 109)
beat_res = {(0, 4): 8, (4, 12): 4}
nb_velocities = 32
additional_tokens = {'Chord': False, 'Rest': True, 'Tempo': False, 'Program': False,
                     'rest_range': (2, 8),  # (half, 8 beats)
                     'nb_tempos': 32,  # nb of tempo bins
                     'tempo_range': (40, 250)}  # (min, max)
tokenizer = REMIEncoding(pitch_range, beat_res, nb_velocities, additional_tokens)
```

```python
tokenizer.vocab.count
```

```python
from tqdm.notebook import tqdm
```

```python
tokens_list = []
```

```python
for i in tqdm(files):
    if i.endswith(".mid"):
        midi = midi = MidiFile("session/"+i)
        if len(tokenizer.midi_to_tokens(midi))==0:
            continue
        tokens_list.append(tokenizer.midi_to_tokens(midi)[0])
```

```python
len(tokens_list)
```

```python
start = int(0x2200)
ans = ''
for tokens in tqdm(tokens_list):
    ans1 = ''
    for i in range(len(tokens)):
        number = start+tokens[i]
        ans1+= chr(number)
    ans+=ans1
    ans+='[SEP]'
```

```python
with open("encoded.txt",'w+',encoding="utf-8") as f:
    f.write(ans)
```
